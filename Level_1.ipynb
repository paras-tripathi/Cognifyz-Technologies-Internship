## Objective
Perform data exploration, preprocessing, descriptive analysis, and geospatial analysis on the restaurant dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

# Load Dataset
df=pd.read_csv('Dataset.csv')
df.head()
```

## Task 1: Dataset Overview

```python
print('Rows, Columns:',df.shape)
df.info()
df.describe(include='all')
```

```python
# Missing Values
missing=df.isnull().sum().sort_values(ascending=False)
missing[missing>0]
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'],bins=20)
plt.title('Distribution of Aggregate Rating')
plt.show()
```

### Observation
- The dataset contains restaurant information including ratings, cuisines, pricing, and location.
- Aggregate Rating is the target variable.
- Missing values are mainly present in text-based columns such as Cuisines.

## Task 2: Descriptive Analysis

```python
display(df.describe())
print(df['Country Code'].value_counts())
print(df['City'].value_counts().head(10))
print(df['Cuisines'].value_counts().head(10))
```

```python
plt.figure(figsize=(10,5))
df['City'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Cities by Restaurants')
plt.show()
```

### Observation
- Identify cities with the highest restaurant count.
- Popular cuisines dominate the dataset.

## Task 3: Geospatial Analysis

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Longitude'],df['Latitude'],alpha=0.4,s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Locations')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Longitude',y='Latitude',hue='Aggregate rating',palette='viridis',legend=False)
plt.title('Location vs Aggregate Rating')
plt.show()
```

## Conclusion
