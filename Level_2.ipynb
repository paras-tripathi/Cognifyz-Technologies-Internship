## Objective
Analyze restaurant services, price ranges and engineer useful features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
df=pd.read_csv('Dataset.csv')
```

## Task 1: Table Booking & Online Delivery

```python
tb=df['Has Table booking'].value_counts(normalize=True)*100
od=df['Has Online delivery'].value_counts(normalize=True)*100
print(tb)
print(od)
print(df.groupby('Has Table booking')['Aggregate rating'].mean())
print(df.groupby('Has Online delivery')['Aggregate rating'].mean())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Has Table booking')
plt.show()
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Has Online delivery')
plt.show()
```

### Observation
Restaurants offering table booking generally have higher average ratings.

## Task 2: Price Range Analysis

```python
print(df['Price range'].value_counts())
print(df.groupby('Price range')['Aggregate rating'].mean())
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df,x='Price range',y='Aggregate rating')
plt.title('Average Rating by Price Range')
plt.show()
```

### Observation
Price range influences customer ratings and service quality.

## Task 3: Feature Engineering

```python
df['Restaurant Name Length']=df['Restaurant Name'].str.len()
df['Address Length']=df['Address'].str.len()
df['Has Table Booking']=df['Has Table booking'].map({'Yes':1,'No':0})
df['Has Online Delivery']=df['Has Online delivery'].map({'Yes':1,'No':0})
df[['Restaurant Name Length','Address Length','Has Table Booking','Has Online Delivery']].head()
```
